```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
progress0 = pd.read_csv('progress0.csv')
progress2 = pd.read_csv('progress2.csv')
progress3 = pd.read_csv('progress3.csv')
progress4 = pd.read_csv('progress4.csv')
progress5 = pd.read_csv('progress5.csv')
progress6 = pd.read_csv('progress6.csv')
progress8 = pd.read_csv('progress8.csv')
progress10 = pd.read_csv('progress10.csv')
progress12 = pd.read_csv('progress12.csv')
progress14 = pd.read_csv('progress14.csv')
progress16 = pd.read_csv('progress16.csv')
progress18 = pd.read_csv('progress18.csv')
progress20 = pd.read_csv('progress20.csv')
```

```python
progress_combined = pd.concat([progress0, progress2, progress3, progress4, progress5, progress6, 
                              progress8, progress10, progress12, progress14, progress16, progress18, progress20], axis = 0)
```

```python
progress_combined.shape
```

```python
plt.plot([i for i in range(progress_combined.shape[0])], progress_combined['episode_reward_mean'])
# plt.title('Episode Reward Mean Across Episodes')
plt.ylabel('Rewards')
plt.xlabel('Iterations')
plt.savefig('episode_reward_mean_ddqn')
plt.show()
```

```python
plt.plot([i for i in range(progress_combined.shape[0])], progress_combined['sampler_results/policy_reward_mean/player_1'])
# plt.title('Episode Reward Mean Across Episodes For Player 1')
plt.ylabel('Rewards')
plt.xlabel('Iterations')
plt.savefig('episode_reward_mean_ddqn_player_1')
plt.show()
```

```python
plt.plot([i for i in range(progress_combined.shape[0])], progress_combined['sampler_results/policy_reward_mean/player_0'])
# plt.title('Episode Reward Mean Across Episodes For Player 0')
plt.ylabel('Rewards')
plt.xlabel('Iterations')
plt.savefig('episode_reward_mean_dqn_player_0')
plt.show()
```

```python
sns.kdeplot(progress_combined['policy_reward_min/player_1'], label = 'Min reward dsitribution')
sns.kdeplot(progress_combined['policy_reward_max/player_1'], label = 'Max reward dsitribution')
plt.xlabel('Reward Distribution of Player 1')
plt.legend()
plt.savefig('episode_reward_dist_ddqn_player_1')
plt.show()
```
